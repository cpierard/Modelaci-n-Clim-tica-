```julia
using DataFrames
using PyPlot
```

```julia
A = readdlm("Antartic_input.txt", '\t')
Antartic_input = DataFrame()
Antartic_input[:x_scale] = A[:,1]
Antartic_input[:Bedrock] = A[:,2]
Antartic_input[:Surface] = A[:,3]
Antartic_input[:Latitude] = A[:,4]
Antartic_input
```

```julia
plot(Antartic_input[:,1],Antartic_input[:,2])
plot(Antartic_input[:,1],Antartic_input[:,3])
plot(Antartic_input[:,1],zeros(length(Antartic_input[:,1])))
xlabel("x scale")
ylabel("Elevation (m)")
title("Antartica actual conditions")
```
